# Import Library

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity, cosine_distances
from sklearn.feature_extraction.text import TfidfVectorizer
import math
```

# Import Hasil Text Preprocessing

```python
df = pd.read_csv("Hasil_Preprocessing.csv",sep=',')
df
```

# Buat Fungsi Algoritma TF-IDF dan Cosine Similarity

```python
def cosine_sklearn(docs):
    # Inisialisasi TF-IDF Vectorizer
    tfidf_vectorizer = TfidfVectorizer(norm=None)

    # Generate TF-IDF vector berdasarkan corpus
    tfidf_matrix = tfidf_vectorizer.fit_transform(docs)
    
    cosine_sim = cosine_similarity(tfidf_matrix)
    return cosine_sim
```

```python
cosine_sklearn = cosine_sklearn(df['cleaned'])
```
