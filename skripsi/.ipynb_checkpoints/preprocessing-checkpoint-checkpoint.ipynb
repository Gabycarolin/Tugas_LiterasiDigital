```python
# Import Library
import re
import pandas as pd
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
# Buat Fungsi Preprocessing

# Cleaning
def cleaning(tweet):
    # Hapus pencatuman username di tweet ex:@jokowi    
    _tweet = re.sub(r"@[\w]*", "", tweet)
    # Hapus hyperlink  
    _tweet =  re.sub(r"https?:\/\/.*[\r\n]*", "", _tweet)
    # Hapus Special Character    
    _tweet = re.sub(r"[()\"#/@;:<>{}*`'+=~|.!?,]", "", _tweet)
    # Remove multiple space 
    _tweet = re.sub("\s\s+", " ", _tweet)
    return _tweet

# Casefolding
def casefolding(tweet):
    return tweet.lower()

# Tokenizing
def tokenizing(tweet):
    return tweet.split()

# Tokenizing
def steeming(tweet):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    return stemmer.stem(tweet)
```

```python
df = pd.read_csv("Hasil_Scrape_Tweet_2021-12-26_23:16-57.csv",sep=',')
df
```

```python
df['casefolding'] = df['text'].apply(cleaning).apply(casefolding)
df['tokenizing'] = df['casefolding'].apply(tokenizing)
df['steeming'] = df['casefolding'].apply(steeming)
```
